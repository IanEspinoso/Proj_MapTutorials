```python

import pandas as pd

df = pd.read_csv('1432sc.csv', sep= ';', skiprows=4, engine='python', skipfooter=2, encoding='utf-8')

df.head(5)
```

```python
# remove last column 
df = df.drop(columns='Unnamed: 45')

# rename first column
df.rename(columns = {'Unnamed: 0': 'community'}, inplace=True)

# visualize first 5 rows
df.head(5)
```

```python
# analyse column types, and missing values
df.info()
```

```python
# convert columns 1975-2018 in type float
for column in map(str,range(1975,2019)):
    df[column] = df[column].str.replace(',','.').astype(float)
```

```python
# check column data types
df.dtypes
```

```python
# denominations in the json file
import json
communities_geo = r'spain-communities.geojson'

# open the json file - json.load() methods returns a python dictionary
with open(communities_geo) as communities_file:
    communities_json = json.load(communities_file)
    
# we loop through the dictionary to obtain the name of the communities in the json file
denominations_json = []
for index in range(len(communities_json['features'])):
    denominations_json.append(communities_json['features'][index]['properties']['name'])
    
denominations_json
```

```python
# names in the data frame
dataframe_names = df.community.tolist()

# names in the json file - the same order as in the data frame 
geojson_names = ['Total Nacional', 'Andalucia', 'Aragon', 'Asturias', 'Baleares', 'Canarias', 'Cantabria',
                 'Castilla-Leon', 'Castilla-La Mancha', 'CataluÃ±a', 'Valencia', 'Extremadura', 'Galicia',
                 'Madrid', 'Murcia', 'Navarra', 'Pais Vasco', 'La Rioja', 'Ceuta', 'Melilla']

# replace data frame names by json names
df.replace(dict(zip(dataframe_names, geojson_names)), inplace=True)
df
```

```python
import folium

# file name - file is located in the working directory
communities_geo = r'spain-communities.geojson' # geojson file

# create a plain world map
communities_map = folium.Map(location=[40.416775, -3.703790], zoom_start=6, tiles='stamenwatercolor')

# generate choropleth map 
communities_map.choropleth(
    geo_data=communities_geo,
    data=df,
    columns=['community', '2018'],
    key_on='feature.properties.name',
    fill_color='YlGnBu', 
    fill_opacity=1, 
    line_opacity=1,
    legend_name='Births per 1000 inhabitants',
    smooth_factor=0)

# display map
communities_map
```
